**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Ваш код здесь
y = data[data['salary']=='>50K']
married_filter = \
(y['marital-status'] == 'Married-civ-spouse') | \
(y['marital-status'] == 'Married-spouse-absent') | \
(y['marital-status'] == 'Married-AF-spouse')
z = len(y[married_filter])/len(y)
z1=(len(y)-len(y[married_filter]))/len(y)
if z>z1: 
    print('Среди много зарабатывающих больше женатых')
if z1>z:
    print ('Среди мног зарабатывающих больше холостых')
```
